# Load Iris dataset

```python
from sklearn import datasets
from sklearn import preprocessing
from sklearn.preprocessing import normalize
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import itertools


def load_iris_dataset(training_size, test_size, feature_dim):
    class_labels = [r'A', r'B', r'C']
    data, target = datasets.load_iris(return_X_y=True)


    #Compute standardisation and normlization only on two classes
    #i.e. filtering class 2, before any computation
    #This is what seems in the paper happens
    selectors = [x is not 1 for x in target]
    data = np.array([data[i] for i,v in enumerate(data) if target[i] != 2])
    target = np.array([x for x in target if x != 2])
    
    # Plot the training points
    colors = ['blue' if t == 0 else 'orange' if t == 1 else 'red' for t in target]
    plt.scatter(data[:, 0], data[:, 1], c=colors)
    
    
    '''
    data = data[:, :feature_dim]
    scaler = StandardScaler().fit(data)
    data = scaler.transform(data)
    data = preprocessing.normalize(data, axis=1)
    '''
    
    sample_train, sample_test, label_train, label_test = \
        train_test_split(data, target, test_size=test_size, random_state=1)
    
    
    sample_train = sample_train[:, :feature_dim]
    sample_test = sample_test[:, :feature_dim]
    

    #Looking at the paper seems it doesn't assume a PCA
    #pca = PCA(n_components=feature_dim).fit(sample_train)
    #sample_train = pca.transform(sample_train)
    #sample_test = pca.transform(sample_test)
    
    
    #Standardization
    # Now we standardize for gaussian around 0 with unit variance
    scaler = StandardScaler().fit(sample_train)
    sample_train = scaler.transform(sample_train)
    sample_test = scaler.transform(sample_test)
    
    
    #Normalization
    sample_train = preprocessing.normalize(sample_train, axis=1)
    sample_test = preprocessing.normalize(sample_test, axis=1)

    
    training_input = {key: (sample_train[label_train == k, :])[:training_size]
                      for k, key in enumerate(class_labels)}
    test_input = {key: (sample_test[label_test == k, :])[:test_size]
                  for k, key in enumerate(class_labels)}
    
    return training_input, test_input
```

## Quantum Circuit

```python
class QDistanceBasedClassifier:
    
####################################################################
    def get_circuit(self):
        return self.qc

####################################################################
    def _qr_ccry(self, theta, a, b, t):
        self.qc.ccx(a,b,t)

        self.qc.cx(b,t)    
        self.qc.ry(theta, t)

        self.qc.cx(b,t)    
        self.qc.ry(-theta, t)

        self.qc.ccx(a,b,t)

        self.qc.cx(b,t)    
        self.qc.ry(-theta, t)

        self.qc.cx(b,t)    
        self.qc.ry(theta, t)
        
####################################################################
        
    def _step_A(self, a, b):
        '''Uniform superpos of ancilla and index qubit'''
        self.qc.h(a)
        self.qc.h(b)

####################################################################

    def _step_B(self, theta, a, c):
        '''Loading test vector  '''
        #qc.cry(theta, a, c)
        self.qc.cx(a, c)
        self.qc.u(-theta, 0, 0, c)  #Seems not equivalent to qc.cry(theta, a, c)
        self.qc.cx(a, c)
        self.qc.u(theta, 0, 0, c)

        self.qc.barrier()
        self.qc.x(a)
        self.qc.barrier()
        
####################################################################

    def _step_C(self, a, b, c):
        '''Loading first training vector'''
        self.qc.ccx(a,b,c)
        self.qc.x(b)
        self.qc.barrier()

####################################################################
        
    def _step_D(self, theta, a, b, c, d):
        '''Load second training vector'''
        self._qr_ccry(theta, a, b, c)
        self.qc.barrier()
        self.qc.cx(b, d) #In paper before this statement qc.swap(c,d). If added --> not properly working
        self.qc.barrier()
        
####################################################################

    def _step_F(self, a):
        '''Compute distance'''
        self.qc.h(a)
        self.qc.barrier()
        
####################################################################

    def step_measure(self):
        '''Measurements'''
        self.qc.measure(self.aqr, self.c_areg)
        #self.qc.measure(self.mqr, self.c_mreg)
        #self.qc.measure(self.iqr, self.c_ireg)#.c_if(c_i, 0)
        self.qc.measure(self.cqr, self.c_creg)
        
####################################################################

    def _init_qc(self):
        aqr = QuantumRegister(1, name="ancilla ")
        mqr = QuantumRegister(1, name="index qregister ") #Just one since we have only two training inputs
        iqr = QuantumRegister(1, name="data register ")
        cqr = QuantumRegister(1, name="classes ")

        c_areg = ClassicalRegister(1, name="classical ancilla ")
        #c_mreg = ClassicalRegister(1, name="classical index ")
        #c_ireg = ClassicalRegister(1, name="classical data ")
        c_creg = ClassicalRegister(1, name="classical classes ")
        
        qc = QuantumCircuit(aqr, mqr, iqr, cqr, c_areg, c_creg)
        return qc, aqr, mqr, iqr, cqr, c_areg, c_creg
    
####################################################################

    def __init__(self, theta_tr_1, theta_test): #Not loading thetha_tr_0
        self.qc, self.aqr, self.mqr, self.iqr, self.cqr, self.c_areg, self.c_creg = self._init_qc()
        
        #Uniform superpos of ancilla and index qubits
        self._step_A(self.aqr, self.mqr)

        #Loading test vector
        self._step_B(theta_test, self.aqr, self.iqr)

        #Loading first training vector
        self._step_C(self.aqr, self.mqr, self.iqr)

        #Load second training vector
        self._step_D(theta_tr_1, self.aqr, self.mqr, self.iqr, self.cqr)

        #compute distance
        self._step_F(self.aqr)
        
        self.step_measure()

        #Measurements are added after the draw in the corrisponding simulator
        #Since unitary_simultor, needs an unitary matrix, measurements need to be postponed        

####################################################################
```

### QASM Simulator

```python
def execute_QASM(qc):
    #q_distance_based_classifier._step_measure(aqr, iqr, mqr, cqr, c_areg, c_ireg, c_mreg, c_creg)
    backend = BasicAer.get_backend('qasm_simulator')
    result = execute(qc, backend, shots=1024).result()
    #plot_bloch_multivector(result.get_statevector())
    return result
```

 ### StateVector Simulator

```python
#step_measure(distance_based_circuit, aqr, iqr, mqr, cqr, c_areg, c_ireg, c_mreg, c_creg)
def execute_statevector(qc):
    backend = BasicAer.get_backend('statevector_simulator')
    result = execute(qc, backend).result()
    statevector = result.get_statevector(qc)
    return statevector
```

### Unitary Simultator 

```python
def retrieve_circuit_as_unitary_matrix(qc):
    simulator = Aer.get_backend('unitary_simulator')
    result = execute(qc, simulator, shots = 1024).result()
    unitary = result.get_unitary(qc)
    print("Circuit unitary:\n", unitary)
    print(unitary[:0])
    return unitary
```

```python
def circuit_from_unitary_matrix(unitary_matrix):
    g = UnitaryGate(unitary_matrix)

    print("Num Qubits:" +str(g.num_qubits))
    print("Num Classical: " +str(g.num_clbits))

    qr = QuantumRegister(4)
    cr = ClassicalRegister(4)

    qc = QuantumCircuit(qr,cr)

    qc.append(g, [qr[0], qr[1], qr[2], qr[3]])

    qc.measure(qr[0], cr[0])
    qc.measure(qr[1], cr[1])
    qc.measure(qr[2], cr[2])
    qc.measure(qr[3], cr[3])

    qc.draw()
    
    return qc
```

## Tomography

```python
def get_tomography(qc, result):
    result_counts = result.get_counts(qc)

    total_samples = sum(result_counts.values())

    # define lambda function that retrieves only results where the ancilla is in the |0> state
    post_select = lambda counts: [(state, occurences) for state, occurences in counts.items() if state[-1] == '0']

    # perform the postselection
    postselection = dict(post_select(result_counts))
    postselected_samples = sum(postselection.values())

    print(f'Ancilla post-selection probability was found to be {postselected_samples/total_samples}')

    retrieve_class = lambda binary_class: [occurences for state, occurences in postselection.items() if state[0] == str(binary_class)]

    prob_class0 = sum(retrieve_class(0))/postselected_samples
    prob_class1 = sum(retrieve_class(1))/postselected_samples

    print(f'Probability for class 0 is {prob_class0}')
    print(f'Probability for class 1 is {prob_class1}')
```
